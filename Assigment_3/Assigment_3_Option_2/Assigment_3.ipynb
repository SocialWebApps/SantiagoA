```python
import os
import csv
from operator import itemgetter
from collections import Counter
from prettytable import PrettyTable

CSV_FILE = os.path.join("resources", 'my_connections_comb.csv')

csvReader = csv.DictReader(open(CSV_FILE), delimiter=',', quotechar='"')

contacts = [row for row in csvReader]

titles = []
for contact in contacts:
    if contact['Job Title'] :
        titles.append(contact['Job Title'])

# Print out a table of titles sorted by frequency
pt = PrettyTable(field_names=['Title', 'Freq'])
pt.align = 'l'
pt.max_width['Title'] = 50
c = Counter(titles)
[pt.add_row([title, freq]) 
 for (title, freq) in sorted(c.items(), key=itemgetter(1), reverse=True) 
     if freq > 0 ]
print pt

print "Total number of titles: " + str(len(titles))
```

```python
import re
import string

split_titles = []
#Goes throught the contact list and applies a re filter to
# remove and split titles at '/' and '&' and ',' and ' and ' in order to get individual titles
for contact in contacts:
    for t in re.split('[&/,]|(?: and )',contact['Job Title'],flags=re.IGNORECASE):
        split_titles.append(t.strip());
        
transforms = {
    'sr': 'Senior',
    'jr': 'Junior',
    'ceo': 'Chief Executive Officer',
    'coo': 'Chief Operating Officer',
    'cto': 'Chief Technology Officer',
    'cfo': 'Chief Finance Officer',
    'vp': 'Vice President',
    'svp': 'Senior Vice President',
    'evp' : 'Executive Vice President',
    'avp' : 'Associate Vice President',
    'mgr': 'Manager',
    'assoc': 'Associate',
    'asst': 'Assistant',
    'cofounder':'Co-Founder',
    'co founder':'Co-Founder',
    'co-founder':'Co-Founder',
    'co owner': 'Co-Owner',
    'co-owner': 'Co-Owner',
    'chairman of the board': 'Chairman',
    'dir':'Director',
    }

# Read in a dictionary of titles and split apart any combined titles like "President/CEO."

def multiple_replace(dict,text):
    
  # Create a regular expression  from the dictionary keys
    for key in transforms.keys():
        regex = re.compile("\\b" + key + "\\b",flags=re.IGNORECASE)
        text = regex.sub(dict[key],text,re.IGNORECASE)
    return text

# Replace common/known abbreviations from the dictionary
titles_trans=[]
for t in split_titles:
    titles_trans.append(multiple_replace(transforms,t))

#-------------------------------------------------------------
    
#Keep track of the number of unique titles
num_unique = 0

# Print out a table of titles sorted by frequency
pt = PrettyTable(field_names=['Title', 'Freq'])
pt.align = 'l'
pt.max_width['Title'] = 50
c = Counter(titles_trans)
for title,freq in sorted(c.items(),key=itemgetter(1),reverse=True):
    if freq > 0:
        pt.add_row([title,freq])
        num_unique = num_unique + 1 # increment the number of unique items
print pt

print "Number of unique titles:" , num_unique
print
print "Bellow is a table showing the frequency of occurance with certain tokens"
print

tokens = []
for title in titles:
    tokens.extend([t.strip(',') for t in title.split()])
pt = PrettyTable(field_names=['Token', 'Freq'])
pt.align = 'l'
c = Counter(tokens)
[pt.add_row([token, freq]) 
 for (token, freq) in sorted(c.items(), key=itemgetter(1), reverse=True) 
     if freq > 1 and len(token) > 2]
print pt
```

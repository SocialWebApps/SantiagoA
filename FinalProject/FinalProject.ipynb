##Reads the current line, the range is taken from the dictionary
####With the range we know what the data represents

```python
def readLine(currentLine):
    for key, value in dic_val.items():  
        pos = value[1]
        description = value[2]
   
        lower = pos[0]
        if(lower > 0):
            lower = lower - 1
            upper = pos[1]
        txt = ''
    
        for x in range(lower, upper):
            txt = txt + currentLine[x]
    
        if("PTDTRACE" == key):
            RACE = txt
        if("GESTFIPS" == key):
            STATE = txt
        if("HEFAMINC" == key):
            INCOME = txt
        if("PEEDUCA" == key):
            EDUCATION = txt
        if("GEDIV" == key):
            DIVISION = txt
 
    return RACE,INCOME,EDUCATION,STATE,DIVISION
```

```python
print ("*"*100)
print "Done...."
def dataArrayCounter(txt,array):
    counted = [] 
    array.sort()
    counter =  Counter(array)
    total = 0
    for x in counter:
        count = counter[x]
        total = total + count
        counted.append((x, count))
    return (txt,total,counted)
print ("*"*100)
```

#Organizes the data into a matrix, easier data matipulation

```python
print ("*"*100) 
print "Organizing data into a matrix, easier data matipulation"

def dataMatrixCreation(workingArray):
    temp = []
    out = []
    
    for x in range(0, len(workingArray)): 
	raw_data = workingArray[x]
	state_data = raw_data[0]     
	if(state_data != "null"):
	    
	    state_tupp = state_data[1]
	    
	    total_tupp = state_data[3]
	    total_tupp = total_tupp[1]
	    #total_tupp = total_tupp[1] #<<<< '0' the dictionary key, '1' the total count
    
	    temp.append((state_tupp[0],total_tupp))        
	else:
	    #check the temp against the mainRow if it is not the same length add the missing element as a 0
	    if(len(temp) != len(mainRow)):
		blank_array = []
		for y in range(0, len(mainRow)):
		    blank_array.append([-1,0])#adding this array prevents an error when we do the final formating, the -1 can be egarded as a place holder	  
		    
		for x in range(0, len(temp)):
		    data = temp[x]
		    isInt =  isinstance( data, int )
    
		    if(isInt == False):		
			for y in range(0, len(mainRow)):
			    if(mainRow[y] == data[0]):
				blank_array.pop(y)
				blank_array.insert(y,data)
		temp = blank_array
	    #Strips the state name used to fill the matrix out. We do not need it here
	    temp_out = []
	    for x in range(0, len(temp)):
		dat = temp[x]
		temp_out.append(dat[1])
	    out.append(temp_out)
	    temp =[]
	    
    
    return out

dataMatrix = dataMatrixCreation(final_data_array_Highest_percentage)

for x in range(0, len(dataMatrix)):
    data = dataMatrix[x]
    if(len(data) < 2):
	dat = data[0]
	
	temp = []
	temp.append(dat)
	temp.append(0)
	dataMatrix[x] = temp

print "Done...."
print ("*"*100)  
```

###Bar Graph Color Generate Function

```python
print ("*"*100)   
print "barGraphColorGenerate Function"
def barGraphColorGenerate(NUM_COLORS):
    cm = plt.get_cmap('gist_stern')
    cm = plt.get_cmap('nipy_spectral')
    
    my_colors = []
    for i in range(NUM_COLORS):
        my_colors.append(cm(1.*i/NUM_COLORS))  # color will now be an RGBA tuple
    return my_colors
print "Done...."
print ("*"*100)  
```

#Plot Bar Graph Function

```python
print ("*"*100)   
print "plotBarGraph Function"
def plotBarGraph(title,DataFrame,COLUMNS,ROW,dict):
    fig, ax = plt.subplots()  
    DataFrame.plot(ax=ax, kind='bar', legend=False,linewidth=0,stacked=False,color=barGraphColorGenerate(len(COLUMNS)))
    ax.set_yticks(dict.keys(),minor=False)
    fig.autofmt_xdate()
    ax.legend(loc='center left', bbox_to_anchor=(1, 0.5)) 
    fig.set_size_inches(18.5,10.5)  
    plt.title(title)
    plt.show()
print "Done...."
print ("*"*100)  
```
